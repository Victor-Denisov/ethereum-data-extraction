```python
# Standard Libraries
import pandas as pd
import numpy as np
import requests
from matplotlib import pyplot 
import seaborn as sns 
from datetime import datetime
```

```python
# read the csv
valhallaDF = pd.read_csv('txndata.csv')
timestamps = pd.read_csv('timestamps.csv')
timestamps = timestamps.drop(columns=['Unnamed: 20'])
```

```python
# pre-processing and cleaning the data
index = ['Transaction Hash', 'Address From', 'Address To', 'NFT ID', 'WETH Value']
valhallaCleanDF = pd.DataFrame(columns=index)
count = 1
temp = []
for i in valhallaDF:
    temp.append(i)
    if (count > 4):
        dfTemp = pd.DataFrame([[temp[0],temp[1],temp[2],temp[3],temp[4]]], columns=index)
        valhallaCleanDF = pd.concat([valhallaCleanDF, dfTemp])
        count = 0
        temp = []
    count = count + 1

#remove the pesky period
for column in valhallaCleanDF:
    valhallaCleanDF[column] = valhallaCleanDF[column].str.split('.').str[0]

valhallaCleanDF
```

```python
timestamp = []
for i in timestamps:
    timestamp.append(datetime.utcfromtimestamp(int(i)).strftime('%Y-%m-%d'))
valhallaCleanDF['date'] = timestamp
valhallaCleanDF
```

```python
valhallaCleanDF['WETH Value'] = valhallaCleanDF['WETH Value'].apply(lambda x: int(x)/1000000000000000000)
valhallaCleanDF
```
